```python
import cv2
from pyzbar.pyzbar import decode
import json
from PIL import Image
import pytesseract
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
# tesseract executable path with respect to google colab
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'


def enhance_image(image_path):
    # Read the image
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply a mild blur to reduce noise slightly
    blurred = cv2.GaussianBlur(image, (3, 3), 0)

    # Apply a mild sharpening kernel
    sharpen_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    sharpened = cv2.filter2D(blurred, -1, sharpen_kernel)

    # Use CLAHE (Contrast Limited Adaptive Histogram Equalization) for contrast enhancement
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced = clahe.apply(sharpened)

    return enhanced
```

```python
#function to read ocr
def ocr_core(image):
    text = pytesseract.image_to_string(image)
    return text

def draw_bounding_boxes(image, decoded_objects):
    for obj in decoded_objects:
        points = obj.polygon
        if len(points) == 4:
            points = np.array(points, dtype=np.int32).reshape((4, 2))
        else:
            hull = cv2.convexHull(np.array([point for point in points], dtype=np.float32))
            points = hull.astype(np.int32).reshape((-1, 1, 2))

        cv2.polylines(image, [points], True, (0, 255, 0), 3)

        text = obj.data.decode('utf-8')
        # Handle cases where points[0] may not have two values
        if len(points[0]) == 2:
            x, y = points[0]
            cv2.putText(image, text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
        else:
            print("Warning: Unable to extract coordinates for text overlay.")

    return image
```

```python
#function to extract and save info
def extract_and_save_info(decoded_objects, ocr_text):
    info = {
        "OCR_Text": ocr_text,
        "Barcodes_QRCodes": []
    }

#specifies object type and data
    for obj in decoded_objects:
        obj_info = {
            "Type": obj.type,
            "Data": obj.data.decode('utf-8')
        }
        info["Barcodes_QRCodes"].append(obj_info)

    return info
```

```python
#function to process image
def process_image(image_path, output_image_path):
    try:
        # Enhance image
        enhanced_image = enhance_image(image_path)

        # Decode barcodes and QR codes
        decoded_objects = decode(enhanced_image)

        # Convert enhanced image to PIL format for OCR
        pil_image = Image.fromarray(cv2.cvtColor(enhanced_image, cv2.COLOR_GRAY2RGB))
        ocr_text = ocr_core(pil_image)

        # Draw bounding boxes on the enhanced image
        enhanced_image_with_boxes = draw_bounding_boxes(enhanced_image, decoded_objects)

        # Save extracted information to a dictionary
        info = extract_and_save_info(decoded_objects, ocr_text)

        # Save the final image with bounding boxes
        cv2.imwrite(output_image_path, enhanced_image_with_boxes)

        return enhanced_image, enhanced_image_with_boxes, info

    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        return None, None, None
```

```python
#main function
def main_single_image(dataset_path, output_path):
    # Get the list of all images in the dataset
    image_paths = [os.path.join(dataset_path, image_name) for image_name in os.listdir(dataset_path) if image_name.lower().endswith(('.png', '.jpg', '.jpeg'))]

    if not image_paths:
        print("No images found in the dataset.")
        return

    # Select the first image for demonstration
    image_path = image_paths[0] #you can selct multiple images like [11-18]
    image_name = os.path.basename(image_path)
    output_image_path = os.path.join(output_path, f"new_image_{image_name}")

    # Process the selected image
    original_image = cv2.imread(image_path)
    enhanced_image, enhanced_image_with_boxes, info = process_image(image_path, output_image_path)

    if enhanced_image is None:
        return

    # Save extracted information to JSON
    output_json_path = os.path.join(output_path, '/content/drive/MyDrive/Output/extracted_info.json')
    with open(output_json_path, 'w') as f:
        json.dump(info, f, indent=4)

    # Display the original and enhanced images using plt.show
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Enhanced Image with Boxes")
    plt.imshow(cv2.cvtColor(enhanced_image_with_boxes, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.show()

    print(f"Processing complete. Enhanced image {image_name}, saved to output directory.")
```
